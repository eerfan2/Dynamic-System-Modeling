**Teori Rantai Markov**

Rantai Markov (Markov Chain) pertama kali dikemukakan oleh matematikawan
berkebangsaan Rusia bernama Andrey Andreevich Markov pada tahun 1906.
Ini merupakan model yang menggambarkan peluang perubahan suatu kondisi (state) berdasarkan kondisi (state) pada proses pengamatan sebelumnya. Penyelesaian rantai Markov menggunakan pendekatan matriks dan probabilitas diagram pohon (probability tree diagram).
Penerapan dalam kehidupan sehari-hari contohnya adalah perubahan cuaca,
perubahan kondisi emosional seseorang, atau perpindahan loyalitas konsumen antara dua merek.

Dalam rantai Markov, probabilitas transisi menggambarkan peluang berpindah dari satu kondisi ke kondisi berikutnya, dan ini dapat dituliskan dalam bentuk matriks probabilitas transisi.

Untuk melakukan simulasi dalam bahasa pemrograman Python, pertama-tama lakukan import library.

```python
from numpy import *
from IPython.display import Math
import sympy as sym
```

Selanjutnya, tampilkan matriks transisi yang berisi probabilitas konsumen bertahan atau berpindah merek, yang menjadi dasar perhitungan di periode berikutnya.

```python
# Aplikasi Model Rantai Markov
# Loyalitas konsumen Apple dan Samsung
C = np.array([[0.7, 0.3], [0.2, 0.8]])
print("Matriks C=")
display(Math(sym.latex(sym.simplify(C))))
```

Untuk kondisi awal, misalkan kita mulai dengan distribusi awal 70% konsumen setia ke Apple dan 30% setia ke Samsung.

```python
# Loyalitas awal konsumen
C0 = np.array([0.7, 0.3])
print("Matriks C0=")
display(Math(sym.latex(sym.simplify(C0))))
```

Pada simulasi ini, kita dapat menghitung distribusi loyalitas konsumen di periode selanjutnya dengan mengalikan matriks kondisi awal dengan matriks transisi.

Prediksi periode berikutnya:

```python
# Prediksi periode 1
C1 = dot(C0, C)
print("Matriks C1 = ")
display(Math(sym.latex(sym.simplify(C1))))
```

Distribusi konsumen di periode selanjutnya dapat dihitung dengan cara mengalikan hasil sebelumnya dengan matriks transisi. Perhitungan ini dapat dilanjutkan untuk periode-periode berikutnya hingga mencapai steady state atau kondisi stabil.

Prediksi periode ke-2:

```python
# Prediksi periode 2
C_2 = dot(C, C)
C2 = dot(C0, C_2)
print("Matriks C2 = ")
display(Math(sym.latex(sym.simplify(C2))))
```

Pada periode kedua, distribusi loyalitas konsumen semakin bergerak menuju nilai stabil, namun masih ada perbedaan kecil antara Apple dan Samsung. Dengan proses perkalian matriks yang berulang, kita melihat bagaimana loyalitas konsumen semakin mendekati keseimbangan, meskipun belum sepenuhnya stabil.

Prediksi periode ke-3:

```python
# Prediksi periode 3
C_3 = dot(C_2, C)
C3 = dot(C0, C_3)
print("Matriks C3 = ")
display(Math(sym.latex(sym.simplify(C3))))
```

Pada periode ketiga, perubahan distribusi semakin kecil, dan kita dapat melihat bahwa tren perpindahan loyalitas konsumen antara kedua merek mulai lebih stabil. Hasil distribusi yang dihitung untuk periode ini semakin mendekati distribusi stabil yang akan tercapai pada periode-periode berikutnya.

Prediksi periode ke-4:

```python
# Prediksi periode 4
C_4 = dot(C_3, C)
C4 = dot(C0, C_4)
print("Matriks C4 = ")
display(Math(sym.latex(sym.simplify(C4))))
```

Pada periode keempat, hasilnya menunjukkan distribusi yang semakin mirip dengan distribusi stabil. Perbedaan antara konsumen yang memilih Apple dan Samsung mulai sangat kecil, menunjukkan bahwa sistem telah bergerak menuju steady state, dengan konsumen memilih berdasarkan probabilitas transisi yang tetap.

Prediksi periode ke-5:

```python
# Prediksi periode 5
C_5 = dot(C_4, C)
C5 = dot(C0, C_5)
print("Matriks C5 = ")
display(Math(sym.latex(sym.simplify(C5))))
```

Pada periode kelima, perubahan dari satu periode ke periode berikutnya sangat kecil, yang mengindikasikan bahwa sistem hampir mencapai steady state. Perubahan kecil ini menunjukkan bahwa loyalitas konsumen sudah mendekati titik keseimbangan, yaitu proporsi yang tetap antara konsumen Apple dan Samsung.

Prediksi periode ke-6 :

```python
# Prediksi periode 6
C_6 = dot(C_5, C)
C6 = dot(C0, C_6)
print("Matriks C6 = ")
display(Math(sym.latex(sym.simplify(C6))))
```

Pada periode keenam dan seterusnya, distribusi loyalitas konsumen mencapai steady state. Proses perubahan antara periode ke periode berikutnya sangat kecil, dan kita dapat melihat bahwa sekitar 67% konsumen setia pada Apple, dan 33% setia pada Samsung. Ini adalah kondisi stabil yang tercapai setelah beberapa periode simulasi, yang menunjukkan bahwa dalam jangka panjang, loyalitas konsumen akan terbagi secara konstan sesuai dengan probabilitas transisi yang ada.

**Hasil Stabil (Steady State)**
